### **Question 1: ROC AUC feature importance**

ROC AUC could also be used to evaluate feature importance of numerical variables.

Let's do that
+ For each numerical variable, use it as score (aka prediction) and compute the AUC with the y variable as ground truth.
+ Use the training dataset for that

If your AUC is < 0.5, invert this variable by putting "-" in front

(e.g. -df_train['engine_hp'])

AUC can go below 0.5 if the variable is negatively correlated with the target variable. You can change the direction of the correlation by negating this variable - then negative correlation becomes positive.

Which numerical variable (among the following 4) has the highest AUC?
+ **balance**
+ **day**
+ **duration**
+ **previous**

```python
numerical_features = ['age', 'balance', 'duration', 'day', 'campaign', 'pdays', 'previous']
numerical_features_scores = data_train[numerical_features]
numerical_features_scores['y'] = y_train.copy()
numerical_features_scores.head()
```

```python
from sklearn.metrics import roc_auc_score

X = numerical_features_scores.drop('y', axis=1)
y = numerical_features_scores['y']

auc_scores = {}

# Loop through each numerical feature in the training set (X)
for column in X.select_dtypes(include=['float64', 'int64']).columns:
    # use the column values as prediction scores
    scores = X[column]

    # Compute the ROC AUC using the scores and the ground truth (y)
    auc = roc_auc_score(y, scores)

    # Check if AUC is < 0.5, and if so, invert the feature
    if auc < 0.5:
        print(f"Inverting feature {column}")
        X[column] = -X[column]
        auc = roc_auc_score(y, X[column])

    # Store the AUC score for this feature
    auc_scores[column] = auc

# Sort the AUC scores from largest to smallest
sorted_auc_scores = {k: v for k, v in sorted(auc_scores.items(), key=lambda item: item[1], reverse=True)}
sorted_auc_scores
```

Apply one-hot-encoding using DictVectorizer and train the logistic regression with these parameters:

**LogisticRegression(solver='liblinear', C=1.0, max_iter=1000)**

```python
numerical = numerical_features.copy()
categorical = categorical_columns = list(data_train.dtypes[data_train.dtypes == 'object'].index)
```

### Question 3: Precision and Recall

Now let's compute precision and recall for our model.

+ Evaluate the model on all thresholds from 0.0 to 1.0 with step 0.01
+ For each threshold, compute precision and recall
+ Plot them

At which threshold precision and recall curves intersect?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score

# Get the predicted probabilities for the positive class
y_val_pred = model.predict_proba(X_val)[:, 1]

# Initialize lists to store precision, recall, and thresholds
precisions = []
recalls = []
thresholds = np.arange(0.0, 1.01, 0.01)

# Loop over each threshold from 0.0 to 1.0
for threshold in thresholds:
    # Binarize predictions: convert probabilities into 0s and 1s based on the threshold
    y_pred = (y_val_pred >= threshold).astype(int)
    
    # Compute precision and recall for the current threshold
    precision = precision_score(y_val, y_pred)
    recall = recall_score(y_val, y_pred)
    
    # Append precision and recall to lists
    precisions.append(precision)
    recalls.append(recall)

# Plot precision and recall as functions of the threshold
plt.figure(figsize=(9, 6))
plt.plot(thresholds, precisions, label='Precision', color='blue')
plt.plot(thresholds, recalls, label='Recall', color='green')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision and Recall at Different Thresholds')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

### Question 4: F1 score

Precision and recall are conflicting - when one grows, the other goes down. That's why they are often combined into the F1 score - a metrics that takes into account both

This is the formula for computing F1:

$$F1 = 2 \cdot \frac{P \cdot R}{P + R}$$

Where P is precision and R is recall.

Let's compute F1 for all thresholds from 0.0 to 1.0 with increment 0.01

At which threshold F1 is maximal?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score

# Get the predicted probabilities for the positive class
y_val_pred = model.predict_proba(X_val)[:, 1]

# Initialize lists to store precision, recall, and thresholds
precisions = []
recalls = []
f1_scores = []
thresholds = np.arange(0.0, 1.01, 0.01)

# Loop over each threshold from 0.0 to 1.0
for threshold in thresholds:
    # Binarize predictions: convert probabilities into 0s and 1s based on the threshold
    y_pred = (y_val_pred >= threshold).astype(int)
    
    # Compute precision and recall for the current threshold
    precision = precision_score(y_val, y_pred)
    recall = recall_score(y_val, y_pred)

    # Calculate F1 score
    if precision + recall > 0:
        f1 = 2 * (precision * recall) / (precision + recall)
    else:
        f1 = 0
    
    # Append precision and recall to lists
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)

# Find the threshold where F1 score is maximal
max_f1 = max(f1_scores)
best_threshold = thresholds[f1_scores.index(max_f1)]

# Print the threshold with the maximal F1 score
print(f"Max F1 score: {max_f1:.3f} at threshold {best_threshold:.2f}")

# Plot F1 score as a function of the threshold
plt.figure(figsize=(10, 6))
plt.plot(thresholds, f1_scores, label='F1 Score', color='red')
plt.xlabel('Threshold')
plt.ylabel('F1 Score')
plt.title('F1 Score at Different Thresholds')
plt.grid(True)
plt.show()
```

### Question 5: 5-Fold CV
Use the KFold class from Scikit-Learn to evaluate our model on 5 different folds:

**KFold(n_splits=5, shuffle=True, random_state=1)**

+ Iterate over different folds of df_full_train
+ Split the data into train and validation
+ Train the model on train with these parameters: **LogisticRegression(solver='liblinear', C=1.0, max_iter=1000)**
+ Use AUC to evaluate the model on validation
  
How large is standard deviation of the scores across different folds?

```python
data_full_train = data_full_train.copy()
data_full_train.head()
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import KFold

# Define your categorical and numerical columns
categorical_columns = list(data_full_train.select_dtypes(include=['object']).columns)
numerical_columns = list(data_full_train.select_dtypes(exclude=['object']).columns)

# Initialize KFold
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# Initialize DictVectorizer
dv = DictVectorizer(sparse=False)

# List to store AUC scores
auc_scores = []

# Iterate over each fold
for train_index, val_index in kf.split(data_full_train):
    # Split the data into training and validation sets
    df_train = data_full_train.iloc[train_index]
    df_val = data_full_train.iloc[val_index]
    
    # Separate the target variable 'y' from the features
    y_train = df_train['y']
    y_val = df_val['y']
    
    # Combine categorical and numerical features for DictVectorizer
    train_dict = df_train[categorical_columns + numerical_columns].to_dict(orient='records')
    val_dict = df_val[categorical_columns + numerical_columns].to_dict(orient='records')
    
    # Transform the features using DictVectorizer
    X_train = dv.fit_transform(train_dict)
    X_val = dv.transform(val_dict)
    
    # Initialize and train the Logistic Regression model
    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000)
    model.fit(X_train, y_train)
    
    # Predict probabilities on the validation set
    y_val_pred_proba = model.predict_proba(X_val)[:, 1]
    
    # Compute AUC for the validation set
    auc = roc_auc_score(y_val, y_val_pred_proba)
    auc_scores.append(auc)

# Convert auc_scores to a numpy array for easier calculation
auc_scores = np.array(auc_scores)

# Calculate the standard deviation of AUC scores
auc_std = np.std(auc_scores)

# Print the standard deviation of AUC scores
print(f"Standard Deviation of AUC scores: {auc_std:.4f}")
```
